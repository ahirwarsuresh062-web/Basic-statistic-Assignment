Q1.What is a Random Variable?

 Ans. A random variable is a variable whose value depends on chance or luck.
  It is used to convert real-life random outcomes into numbers so that we can apply mathematics.

Example:

Toss a coin ‚Üí Head = 1, Tail = 0

Roll a dice ‚Üí numbers from 1 to 6

So, the number we get after an experiment is called a random variable.

Q6. Discrete Uniform Distribution

Ans. In this distribution, all outcomes have equal probability.

Example:

Tossing a fair coin

Rolling a fair dice

Each outcome has the same chance of occurring.

Q7. Bernoulli Distribution (Properties)

Ans. Bernoulli distribution deals with only one trial.

Key points:

Only two outcomes ‚Üí success or failure

Probability of success = p

Probability of failure = 1 ‚àí p

Example:

Pass or Fail

Head or Tail

Q8. Binomial Distribution

Ans. Binomial distribution is used when Bernoulli trials are repeated.

Conditions:

Fixed number of trials

Each trial has two outcomes

Probability remains same

Example:
Tossing a coin 10 times and counting heads.

Q9. Poisson Distribution

Ans. Poisson distribution measures how many times an event occurs in a fixed time or area.

Applications:

Number of accidents per day

Number of calls in a call center

It is used when events occur randomly.

Q10. Continuous Uniform Distribution

Ans. Here, all values in a given range have equal probability.

Example:
Choosing a random number between 0 and 1.

Q11. Characteristics of Normal Distribution
